```python
import requests
res = requests.get(
    'http://ggoreb.com/dinosaur/info.jsp'
)
```

```python
# 문자를 LIst와 Dictionary로 변경
result = res.json() 
```

```python
for re in result: #for - each
    print(re.get("kind"))
```

```python
# 3번째 요소 마다 줄바꿈
for index in range(len(result)):
    print(result[index].get("kind"), end = ' ')
    if (index + 1) % 3 == 0 :
        print()
#     print(type(result[i]))
#     for k, v in result[i].items():
#         print(k)
```
